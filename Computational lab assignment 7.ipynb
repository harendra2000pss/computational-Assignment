```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
new_data = np.loadtxt('Age_Pred.csv',delimiter=',',skiprows=1)
print(new_data)
```

```python
T_data = new_data.T
print(T_data)
```

```python
Age = T_data[0]    #Y attribute is Age
Rbp= T_data[1]     #X1 attribute is Rest blood pressure
Mhr = T_data[2]   #X2 attribute is Maximum heart rate
print('Age :',Age,\
      '\nRbp :',Rbp,\
      '\nMhr :',Mhr)
```

```python
Age=np.array(Age)
Rbp=np.array(Rbp)
Mhr=np.array(Mhr)
n=len(Age)
Age_new=Age.reshape(n, 1)
Rbp_new=Rbp.reshape(n, 1)
Mhr_new=Mhr.reshape(n, 1)
```

```python
plt.xlabel('Rbp',color ='red')
plt.ylabel('Age',color ='red')
plt.title('Rbp vs Age Scatter plot')
plt.scatter(Rbp, Age, color='red')
plt.show()
```

```python
plt.xlabel('Mhr',color ='red')
plt.ylabel('Age',color ='red')
plt.title('Mhr vs Age Scatter plot')
plt.scatter(Mhr, Age, color='green')
plt.show()
```

```python
# trying to visualise X1 and X2 using scattter plot  and predicting the correlation between them 
plt.xlabel('Rbp',color ='red')
plt.ylabel('Mhr',color ='red')
plt.title('Rbp vs Mhr Scatter-plot')
plt.scatter(Rbp, Mhr)
plt.show()
```

```python
#  y = b0 + x1 * b1 + x2 * b2 
# changing our x1 and x2 attributes in 1-d array for manipulation
x_0 = np.ones([n, 1])
print(x_0)
```

```python
Rbp_norm = np.array([])
Mhr_norm = np.array([])
Age_norm = np.array([])
X1 = 0
X2 = 0
X3 = 0
for i in range(Age.shape[0]):
    X1 = (Rbp[i]-np.mean(Rbp))/np.std(Rbp)
    X2 = (Mhr[i]-np.mean(Mhr))/np.std(Mhr)
    X3 = (Age[i]-np.mean(Age))/np.std(Age)
    Rbp_norm = np.append(Rbp_norm, X1)
    Mhr_norm = np.append(Mhr_norm, X2)
    Age_norm = np.append(Age_norm, X3)
    X1 = 0
    X2 = 0
    X3 = 0
    
Rbp_new = Rbp_norm.reshape(n, 1)
Mhr_new = Mhr_norm.reshape(n, 1)
```

```python
# making a single X matrix from two 1-d array 
new = np.hstack((x_0, Rbp_new, Mhr_new)) 
print('\t\t X matrics')
print()
print(new)
new_T = new.T
print('\n\n\n')
print('\t\t\t  Transpose of X  \n\n')
print(new_T)
```

```python
# X.T * X
new_T_new = new_T.dot(new)
print('\t\t   X_Transpose*X \n\n')
print(new_T_new)
print('\n\n')

# correlation matrix to detect multicollinearity
b = new.shape[0]
correlation_matrix  = new_T_new/b 
print('\t\t  Correlation matrix  \n\n')
print(correlation_matrix )
print('\n\n')

# inverse of ((X.T)X)
inverse_new= np.linalg.inv(new_T_new)
print('\t  Inverse of X Transpose*X matrix  \n\n')
print(inverse_new)
```

```python
# calculating (X.T)Y
new_T_Age = (new_T).dot(Age_new)
print('\t\t  X-transpose*Y matrix \n\n')
print(new_T_Age)
```

```python
# calculating for beta
betas = inverse_new.dot(new_T_Age)
print("beta matrix is  \n\n", betas)

beta_0 = betas[0]
beta_1 = betas[1]
beta_2 = betas[2]
print('\n\nValue of intercept, beta1, beta2 \n')
print("beta0 = ", beta_0, "\nbeta1 = ", beta_1, "\nbeta2 = ", beta_2)

# fitting value 
print('\n\n\t     Our fitted line for model')
print()
print('Y =', beta_0, '+', beta_1, 'X1','+', beta_2, 'X2')
```

```python
# findind Y-pred
Age_pre = np.array([])
Age_pre1 = 0
for i in range(new.shape[0]):
    for j in range(new.shape[1]):
        Age_pre1 = Age_pre1 + (betas[j] * new[i][j])
    Age_pre = np.append(Age_pre, Age_pre1, axis = 0)
    Age_pre1 = 0

Age_pre = Age_pre.reshape(n,1)
print('y-hat :\n\n', Age_pre)
```

```python
# our scatter plot in 3d
fig = plt.figure(figsize=(8,8))
new_fig = fig.add_subplot(111, projection='3d')
   
new_fig.scatter(new[:,1], new[:,2], Age_new, color='skyblue',label ='actual cement' )
new_fig.scatter(new[:,1], new[:,2], Age_pre, c ='orange',label ='predicted cement')
new_fig.set_xlabel('Rbp')
new_fig.set_ylabel('Mhr')
new_fig.set_zlabel('Age')
plt.title('Scatter plot of Age, Rbp, Mhr')
new_fig.legend()
new_fig.view_init(45, 0)
plt.show()
```

```python
%matplotlib notebook
xs = np.tile(np.arange(3),(3,1))
ys = np.tile(np.arange(3),(3,1)).T
z = xs*beta_1 +ys*beta_2 + beta_0

fig = plt.figure(figsize=(8,8))
new_fig = fig.add_subplot(111, projection='3d')
new_fig.scatter(new[:,1], new[:,2], Age_new, color='skyblue',label ='actual cement',alpha = 0.3)
new_fig.scatter(new[:,1], new[:,2], Age_pre, c ='orange',label ='predicted cement',alpha = 0.3)
new_fig.plot_surface(xs, ys, z,alpha = 1.0)
new_fig.set_xlabel('Rbp')
new_fig.set_ylabel('Mhr')
new_fig.set_zlabel('Age')
plt.title('Scatter plot of Age, Rbp, Mhr')
new_fig.legend()
new_fig.view_init(45, 0)
plt.show()
```

```python
# calculting R-square value
numerator = 0.0
denomenator = 0.0

Age_mean = np.mean(Age)

for i in range(len(Age)):
    numerator = numerator + ((Age_pre[i] - Age_mean)**2)
    denomenator = denomenator + ((Age[i] - Age_mean)**2)

R_square = (numerator/denomenator)
print('R-square value is : ', R_square)  
```

```python
#anova attributes
ss_reg = 0.0
ss_res = 0.0
for i in range(new.shape[0]):
    ss_reg += (Age_pre[i]-Age_mean)**2
    ss_res += (Age[i]-Age_pre[i])**2

print("SS_regression :", ss_reg, "\nSS_residual:", ss_res,"\nSS_total :", ss_reg+ss_res)

print('\n\n')

n = new.shape[0]
ms_reg = (ss_reg/2)
ms_res = (ss_res/(n-3))
print("MS Regression :", ms_reg, "\nMS Residual :", ms_res)

print('\n\n')

F_stats = (ms_reg/ms_res)
print("F-stats : ", F_stats)
std_Er = (ms_res**0.5) 
print("Std Err :", std_Er)
```

```python
# anova table
F_significant = 2.33E-17 
print("ANOVA\n","\t\t","degree of freedom\t\t", "SS\t\t\t", "MS\t\t", "  F-stats\t", "significance-F")
print("\nRegression\t", "k-1=", 2 , "\t\t", ss_reg, "\t", ms_reg, "\t", F_stats, "\t  ", F_significant)
print("\nResidual\t", "n-k =", 296, "\t\t", ss_res, "\t", ms_res )
print("\nTotal\t\t", "n-1 =", 298, "\t\t", ss_reg+ss_res )
```
